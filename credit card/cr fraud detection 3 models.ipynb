```python
#Model building
from sklearn.model_selection import train_test_split
```

```python
# Model 1 : Random forest classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score
from sklearn.metrics import classification_report
X=data.drop(['Class'],axis=1)
y=data['Class']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=123)
rfc=RandomForestClassifier()
model=rfc.fit(X_train,y_train)
prediction1=model.predict(X_test)
a1=accuracy_score(y_test,prediction1)
```

```python
classification_report(y_test,prediction1)
```

```python
f1=f1_score(y_test,prediction1)
f1
```

```python
pd.crosstab(y_test,prediction1)
```

```python
print('Accuracy of model1 is',a1)
print('f1_score of model1 is',f1)
```

```python
# cross-validation
from sklearn.model_selection import cross_val_score
print(cross_val_score(model,X,y,cv=5,scoring = 'accuracy').mean())
```

```python
# Model 2 : Logistic Regression
from sklearn.linear_model import LogisticRegression
X1=data.drop(['Class'],axis=1)
y1=data['Class']
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.30,random_state=123)
lr=LogisticRegression()
model2=lr.fit(X1_train,y1_train)
prediction2=model2.predict(X1_test)
a2=accuracy_score(y1_test,prediction2)
```

```python
classification_report(y1_test,prediction2)
```

```python
pd.crosstab(y1_test,prediction2)
```

```python
f2=f1_score(y1_test,prediction2)
```

```python
print('Accuracy of model2 is',a2)
print('f1_score of model2 is',f2)
```

```python
# cross-validation
from sklearn.model_selection import cross_val_score
print(cross_val_score(model2,X1,y1,cv=5,scoring='accuracy').mean())
```

```python
# Model 3 : Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
X2=data.drop(['Class'],axis=1)
y2=data['Class']
X2_train,X2_test,y2_train,y2_test=train_test_split(X2,y2,test_size=0.3,random_state=123)
dr=DecisionTreeRegressor()
model3=dr.fit(X2_train,y2_train)
prediction3=model3.predict(X2_test)
a3=accuracy_score(y2_test,prediction3)
```

```python
classification_report(y2_test,prediction3)
```

```python
pd.crosstab(y2_test,prediction3)
```

```python
f3=f1_score(y2_test,prediction3)
```

```python
print('Accuracy of model3 is',a3)
print('f1_score of model3 is',f3)
```

```python
#cross-validation
from sklearn.model_selection import cross_val_score
print(cross_val_score(model3,X2,y2,cv=5,scoring='accuracy').mean())
```
